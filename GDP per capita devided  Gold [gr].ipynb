## -------------------------------------import libraries ----------------------------------------

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## -------------------------------------read data -------------------------------------------------

```python
data = pd.read_csv(r"C:\Users\LENOVO\Desktop\Book2.csv")
data["Country Name"]
```

```python
# show data
data2 = data.set_index("Country Name")
data2.head()
```

## -------------------------------------main calc -------------------------------------------------

```python
# columns must be int 
x = data.columns[1:].astype(int)
```

```python
#polotting for United Arab Emirates
plt.plot(x, data2.loc["United Arab Emirates"] ,color='b', label='UEA', linewidth ='3')
plt.ylabel("GDP per gold  [gr]")
plt.xlabel("Time")
plt.title("gdp per capita/ gold [gr] for United Arab Emirates over time")
plt.rcParams["figure.figsize"] = [10, 5]
plt.xlim(1970, 2023)
plt.grid()
plt.legend(title = "country: ") 
plt.show()
```
